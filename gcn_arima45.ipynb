```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.data import Data

# Define the graph
edge_index = torch.tensor([[0, 1, 1, 2], [1, 0, 2, 1]], dtype=torch.long)  # Edges
x = torch.tensor([[1, 0], [0, 1], [1, 1]], dtype=torch.float)  # Node features
data = Data(x=x, edge_index=edge_index)

# Define the 2-layer GCN model
class GCN(torch.nn.Module):
    def __init__(self):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(2, 2)  # Input dim=2, output dim=2
        self.conv2 = GCNConv(2, 2)  # Input dim=2, output dim=2

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)  # Apply ReLU activation
        x = self.conv2(x, edge_index)
        return F.relu(x)  # Apply ReLU activation

# Initialize and run the model
model = GCN()
output = model(data.x, data.edge_index)
print("Output features:\n", output)
```

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA

# Define the time series data
data = [10, 12, 11, 15, 14, 16, 18, 20, 19, 22]

# Fit the ARIMA model
model = ARIMA(data, order=(1, 0, 1))  # ARIMA(p=1, d=0, q=1)
model_fit = model.fit()

# Print the model summary
print(model_fit.summary())

# Forecast the next value
forecast = model_fit.forecast(steps=1)
print("Forecasted value:", forecast[0])
```
